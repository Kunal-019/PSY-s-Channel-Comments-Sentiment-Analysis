```python
import pandas as pd
import numpy as np
import nltk
import warnings

%matplotlib inline    

warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('/content/Youtube01-Psy.csv')
```

```python
dataset
```

```python
from nltk.tokenize import word_tokenize
```

```python
def tokenize(comment):
    return word_tokenize(comment)
```

```python
dataset['tokens'] = dataset['CONTENT'].apply(tokenize)
```

```python
print('Tokens for first comment:')
print(dataset['tokens'][0])
```

```python
new_dataset = dataset[['CONTENT', 'CLASS']]
```

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
new_dataset
```

```python
# removes pattern in the input text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
# remove twitter handles (@user) and store it in new column clean_text
new_dataset['CONTENT'] = np.vectorize(remove_pattern)(new_dataset['CONTENT'], "@[\w]*")
```

```python
# remove special characters, numbers and punctuations
new_dataset['CONTENT'] = new_dataset['CONTENT'].str.replace("[^a-zA-Z#]", " ")
new_dataset.head()
```

```python
# remove short words

new_dataset['CONTENT'] = new_dataset['CONTENT'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))
new_dataset.head()
```

```python
#lower case
new_dataset['CONTENT'] = new_dataset['CONTENT'].apply(lambda x: x.lower())
```

```python
new_dataset
```

```python
import nltk
nltk.download('all')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
     
```

```python

from textblob import TextBlob

def get_sentiment(comment):
    blob = TextBlob(comment)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'


dataset['sentiment'] = dataset['CONTENT'].apply(get_sentiment)

sentiment_counts=print(dataset['sentiment'].value_counts())
```

```python
print('\nSample comments by sentiment:')
for sentiment in ['positive', 'negative', 'neutral']:
    sample = dataset[dataset['sentiment'] == sentiment]['CONTENT'].sample(5, random_state=1)
    print(f'{sentiment.capitalize()} comments:\n{sample}\n')
```

data tranformation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1420)
```

```python
X = cv.fit_transform(new_dataset['CONTENT']).toarray()
y = dataset.iloc[:, -1].values
```

```python
# Saving BoW dictionary to later use in prediction
import pickle
bow_path = 'c1_BoW_Sentiment_Model.pkl'
pickle.dump(cv, open(bow_path, "wb"))
```

### Dividing dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

### Model fitting (Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Exporting NB Classifier to later use in prediction
import joblib
joblib.dump(classifier, 'c2_Classifier_Sentiment_Model') 
```

###Model performance


```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy_score(y_test, y_pred)
```
